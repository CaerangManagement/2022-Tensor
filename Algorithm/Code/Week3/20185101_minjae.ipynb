```python
exam_num = int(input())

tester = list(map(int, input().split()))

main_sv, sub_sv = map(int, input().split())

sv_num = 0

tester = [tester[i]-main_sv for i in range(len(tester))]
sv_num += len(tester)


for i in tester:
  if i <= 0:
    continue
  else:
    if i%sub_sv:
      sv_num+=(i//sub_sv+1)
    else:
      sv_num+=(i//sub_sv)
print(sv_num)

```
