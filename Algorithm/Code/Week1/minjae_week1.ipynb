```python
from decimal import Decimal
a, b = map(int, input().split())
print('{:.20f}'.format((Decimal(a)/Decimal(b)))) # 정확한 수 표현 위해 Decimal 사용
```

```python
#1+2+3+4+... 일때 n이 최대
S = int(input())
n=1
while (n*(n+1)/2<=S):
  n+=1
print(n-1)


```

```python
size = int(input())


filter = [list(map(int,input().split())) for i in range(size) ]

while len(filter) != 1: #filter의 크기가 1이 될때 종료
  new_filter=[] #풀링을 했을때 생성된 필터
  for i in range(0,len(filter),2):
    pooled=[] #new_filter를 2번째로 큰값들로 채우기 위해 만든 리스트
    for j in range(0,len(filter),2):
      solted_filter = sorted([filter[i][j], filter[i][j+1], filter[i+1][j], filter[i+1][j+1]]) #2x2의 요소들을 비교해서 내림차순 정렬
      pooled.append(solted_filter[2]) #2번째로 큰값을 pooled에 하나씩 추가
    new_filter.append(pooled) #새로운 필터에 pooled 값들 추가
  filter=new_filter #사이즈가 1이될때 까지 해야하기 때문에 pooling된 필터를 원본 필터에 덮어씌우고 다시 반복 한번 pooling할때마다 사이즈는 1/2
print(filter[0][0])
```
